##### Uniform Distribution

* Description: A random variable where all of outcomes are equally likely.
* Kind: Can be both continuous or discrete
* Example: Rolling a fair, six-sided die.
* Link: [Continous](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.uniform.html#scipy.stats.uniform), [Discrete](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.randint.html)

##### Bernouli Distribution

* Description: A distribution describing a single event with only two possible outcomes.
* Kind: Discrete
* Example: A single coin landing on heads or tails.
* Link: https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.bernoulli.html

```python
from scipy.stats import bernoulli

fig, ax = plt.subplots(1, 1)
p = 0.6
x = np.arange(bernoulli.ppf(0.01, p), bernoulli.ppf(0.99, p))
ax.plot(x, bernoulli.pmf(x, p), 'ro', ms=8, label='bernoulli pmf')
ax.vlines(x, 0, bernoulli.pmf(x, p), colors='r', lw=5, alpha=0.5)
rv = bernoulli(p)
ax.vlines(x, 0, rv.pmf(x), colors='k', linestyles='-', lw=1, label='frozen pmf')
ax.legend(loc='best', frameon=False)
plt.show()
```

##### Binomial Distribution

* Description: A distribution describing an event with two possible outcomes being repeated several times in a row.
* Kind: Discrete
* Example: 10 trials of picking a red or blue marble (and replacing it) out of a bag with only red and blue marbles in it.
* Link: https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.binom.html

```python
from scipy.stats import binom

fig, ax = plt.subplots(1, 1)
n, p = 10, 0.7
mean, var, skew, kurt = binom.stats(n, p, moments='mvsk')
x = np.arange(binom.ppf(0.01, n, p), binom.ppf(0.99, n, p))
ax.plot(x, binom.pmf(x, n, p), 'bo', ms=8, label='binom pmf')
ax.vlines(x, 0, binom.pmf(x, n, p), colors='g', lw=5, alpha=0.5)
rv = binom(n, p)
ax.vlines(x, 0, rv.pmf(x), colors='k', linestyles='-', lw=1, label='frozen pmf')
ax.legend(loc='best', frameon=False)
plt.show()
```

##### Poisson Distribution

* Description: A distribution describing how likely an event is to occur.
* Kind: Discrete
* Example: Finding the probability that more than 10 students in a class will get a 100 on the exam.
* Link: https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.poisson.html

```python
from scipy.stats import poisson

fig, ax = plt.subplots(1, 1)
mu = 0.3
mean, var, skew, kurt = poisson.stats(mu, moments='mvsk')
x = np.arange(poisson.ppf(0.01, mu), poisson.ppf(0.99, mu))
ax.plot(x, poisson.pmf(x, mu), 'bo', ms=8, label='poisson pmf')
ax.vlines(x, 0, poisson.pmf(x, mu), colors='b', lw=5, alpha=0.5)
rv = poisson(mu)
ax.vlines(x, 0, rv.pmf(x), colors='k', linestyles='-', lw=1, label='frozen pmf')
ax.legend(loc='best', frameon=False)
plt.show()
```

##### Normal Distribution

* Description: A distribution for an independent random variable that is also a real number.
* Kind: Continuous
* Example: Height measurements for a particular age group of people.
* Link: https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html

```python
from scipy.stats import norm

fig, ax = plt.subplots(1, 1)
mean, var, skew, kurt = norm.stats(moments='mvsk')
x = np.linspace(norm.ppf(0.01), norm.ppf(0.99), 100)
ax.plot(x, norm.pdf(x), 'b-', lw=5, alpha=0.6, label='norm pdf')
rv = norm()
ax.plot(x, rv.pdf(x), 'k-', lw=2, label='frozen pdf')
vals = norm.ppf([0.001, 0.5, 0.999])
np.allclose([0.001, 0.5, 0.999], norm.cdf(vals))
r = norm.rvs(size=1000)
plt.show()
```

#### Q4.4)

We wish to know if the data is now normally distributed. You can use a Shapiro Wilks test in `scipy.stats` to determine if a distribution is normal. This is demonstrated below.

```python
# Mean of 50, standard deviation of 10, create 100 numbers
normal_values = stats.norm.rvs(loc=50, scale=10, size=100)
plt.hist(normal_values)
plt.title("Normal Distribution with Mean of 50")
plt.ylabel("Frequency")
plt.show()

test_statistic, p_value = stats.shapiro(normal_values)
print("Normally Distributed Values:", test_statistic, p_value, p_value < .05)
print("Since the pvalue is greater than 0.05, we accept the null hypothesis that it is normally distributed.")

# Mean of 50, standard deviation of 10, create 100 numbers
normal_values = stats.randint.rvs(low=0, high=100, size=100)
plt.hist(normal_values)
plt.title("Random Distribution")
plt.ylabel("Frequency")
plt.show()

test_statistic, p_value = stats.shapiro(normal_values)
print("Random Values:", test_statistic, p_value, p_value < .05)
print("Since the pvalue is less than 0.05, we reject the null hypothesis that it is normally distributed.")
```

##### f_oneway

    * Population is normally distributed
    * Populations' standard deviations are equal
* Example: Take a big group of students, break them up by gender, and see if their heights differ by gender.

##### pearsonr

* Function: `pearsonr`
* Link: https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.pearsonr.html
* Test: Pearson Correlation
* Description: Measures linear relationship between two datasets
* Assumptions:
    * Each dataset is normally distributed
    * Variables are continuous
    * Linearity - straight line relationship between each of two variables (equally distributed about regression line)
* Example: Measuring time an airplane is in the air against amount of fuel left  (if air duration increases, fuel level decreases)

##### spearmanr

* Function: `spearmanr`
* Link: https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.spearmanr.html#scipy.stats.spearmanr

* Description: Measures strength of monotonic relationship between two variables; Nonparametric measure of strength and direction of two variable's association.
* Assumptions:
    * Two variables that are ordinal, interval, or ratio
    * The two variables are paired observations
    * There is a monotonic relationship between the two variables (relationship where if value of one variable goes up, so does other variable, or relationship where as one value of variable increases, the other decreases)

* Description: Measures relationship between a binary variable and a continuous variable
* Assumptions:

    * Continuous variable is approximately normally distributed for each category of bianry variable

* Example: A study attempting to determine if there is a relationship between gender and time it takes to complete a survey.

##### kendalltau

##### ttest_ind

* Description: Two sided t-test for means of two independent samples to check if samples have identical or different mean values.
* Assumptions:
    * Populations have identical variances
    * Independence in samples

##### ttest_rel

* Description: Two sided t-test on two related samples to test if they have the same or different mean values.
* Assumptions:

    * Dependent variable must be continuous
    * Related samples (subjects in each sample are the same)

* Example: A study on the effectiveness of a new medicine comparing pain levels before and after using the medication.

##### chisquare

* Example: A study to determine if the relationship between race and the cities in a county people live in is reflective of the population as a whole.

##### mannwhitneyu

* Function: `mannwhitneyu`
* Link: https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.mannwhitneyu.html#scipy.stats.mannwhitneyu
* Test: Mann-Whitney rank test on samples x and y

    * Two samples are independent, unpaired
    * Data is ordinal/continuous 
    * Samples have similar shaped distributions (does *not* have to be normal)
* Example: A study to determine which two tire brands are better for driving in the snow.  Each brand is rated by different users on a scale of 1-10 to determine which tire is better.

##### kstest

* Function: `kstest`
* Link: https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.kstest.html#scipy.stats.kstest
* Test: Kolmogorov-Smirnov test for goodness of fit (one or two samples)
* Description: Performs test of distribution of an observed random variable against another(known) to see whether the distributions are identical.
* Assumptions:
    * Simple random sample
    * Distributions are continuous
    * If one sample, theoretical sample must be known/specified
    * If two sample, samples are mutually independent
* Example: Determining the distributions of local runners from two different generations and the times it took them to run a marathon.

##### wilcoxon

* Function: `wilcoxon`
* Link: https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.wilcoxon.html#scipy.stats.wilcoxon
* Test: Wilcoxon signed-rank test
* Description: Tests whether two related paired samples come from the same distribution.  Nonparametric version of the paired T-Test.
* Assumptions:
    * Paired/dependent samples
    * Randomly and independely chosen paired data
    * Differences are symmetric, mutually independent, with same median
    * Continuous/ordinal data measurements
        * Observations must be comparable so it is clear which one is greater
    * 
* Example: A study assesing the effectiveness of speech therapy for children where a number of children are observed reading a script and the number of times they mispronounce a word is measured before and after speech therapy.  

##### kruskal

* Function: `kruskal`
* Link: https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.kruskal.html#scipy.stats.kruskal
* Test: Krustakl-Walkkish H-test (independent samples)
* Description: Tests whether the median of all groups are equal
* Assumptions:
    * Number of samples in each group must be >= 5
    * Samples are independent
    * Ordinal, ratio, or interval data measurements
    * Distributions for all groups should be the same shape
* Example: A study to determine whether exam performance differs based on hours of study.

#### Q6.5)

Previously, you were asked to calculate whether there was a significant difference between `New York City` and `Chicago` skyscrapers. Now provide a measure of effect size to fully characterize their difference.

```python
from math import sqrt

nyc = skyscrapers.where(skyscrapers['location.city'] == "New York City").dropna()['statistics.height']
nyc = nyc.where(nyc > 0).dropna()

chicago = skyscrapers.where(skyscrapers['location.city'] == "Chicago").dropna()['statistics.height']
chicago = chicago.where(chicago > 0).dropna()

# Calculating Cohen's d
# Reference: https://stackoverflow.com/questions/21532471/how-to-calculate-cohens-d-in-python
# Reference: https://en.wikipedia.org/wiki/Effect_size#Cohen's_d
(nyc.mean() - chicago.mean()) / (sqrt((nyc.std() ** 2 + chicago.std() ** 2) / 2))
```

The measure of effect, 0.14 (Cohen's d), is considered to be small.
